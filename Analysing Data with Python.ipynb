```python
import pandas as pd
```

### Mengimport File Data

```python
data_uncleaned = pd.read_csv('Data_Uncleaned.csv')
data_uncleaned
```

### Menampilkan nilai statistik

```python
data_uncleaned.describe()
```

### Jumlah nilai missing pada setiap kolom

```python
data_uncleaned.isna().sum()
```

### Mengisi nilai missing dengan value angka 0

```python
data_uncleaned.fillna(0,inplace=True)
data_uncleaned
```

### Menghapus nilai duplikat

```python
data_uncleaned.drop_duplicates(inplace=True)
data_uncleaned
```

### Mengubah semua nama kolom menjadi uppercase

```python
data_uncleaned.columns = data_uncleaned.columns.str.upper()
data_uncleaned
```

```python
# identifikasi kolom data yang tidak konsisten nilai datanya
# ubah sesuai type data kolom
```

```python
# mengubah nilai data tidak konsisten
# pada kolom DIRECTOR_FACEBOOK_LIKES menjadi integer

data_uncleaned.at[4,'DIRECTOR_FACEBOOK_LIKES'] = 475
data_uncleaned
```

```python
# mengubah type data kolom DIRECTOR_FACEBOOK_LIKES menjadi integer

data_uncleaned['DIRECTOR_FACEBOOK_LIKES'] = data_uncleaned[
    'DIRECTOR_FACEBOOK_LIKES'].astype(int)
```

```python
# mengubah type data kolom MOVIE_TITLE menjadi string

data_uncleaned['MOVIE_TITLE'] = data_uncleaned[
    'MOVIE_TITLE'].astype(str)
```

```python
# menghapus penulisan berulang pada data
# di kolom MOVIE_TITLE

data_uncleaned['MOVIE_TITLE'] = data_uncleaned[
    'MOVIE_TITLE'].str.replace('?ÿ', '')
data_uncleaned
```
