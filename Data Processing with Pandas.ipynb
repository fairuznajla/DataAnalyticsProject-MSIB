```python
import pandas as pd
import numpy as np
```

```python
foodsales = pd.read_csv('FoodSales.csv')
```

```python
# 1
# Filter data tersebut untuk hanya mengambil category Snacks, tampilkan hanya 3 baris data teratas diurutkan berdasarkan kolom Product (gunakan loc atau iloc)

snacks = foodsales.loc[foodsales.Category=='Snacks']
snacks_sorted = snacks.sort_values('Product', ascending=True).iloc[:3]
snacks_sorted
```

```python
# 2 (tanpa meng-groupby product)
# Berapa selisih penjualan Product yang paling tinggi quantity penjualannya dengan yang paling rendah quantity penjualannya di city Boston?

penjualan_boston = foodsales[foodsales.City=='Boston']
selisih_quantity = penjualan_boston['Quantity'].max() - penjualan_boston['Quantity'].min()
selisih_quantity
```

```python
# 2 (dengan meng-groupby product didapatkan selisih quantity tiap product)
# Berapa selisih penjualan Product yang paling tinggi quantity penjualannya dengan yang paling rendah quantity penjualannya di city Boston?

penjualan_boston = foodsales[foodsales.City=='Boston']
boston_product = penjualan_boston.groupby('Product', as_index=False).agg(quantity_max=('Quantity','max'),
                                                                          quantity_min=('Quantity','min'))
boston_product['SelisihQuantity'] = boston_product['quantity_max']-boston_product['quantity_min']
boston_product
```

```python
# 3
# Buatlah kolom baru untuk mengelompokkan UnitPrice produk menjadi mahal jika nilainya diatas rata-rata UnitPrice dan murah jika dibawah atau sama dengan rata-rata UnitPrice

mean_UnitPrice = foodsales['UnitPrice'].mean()
foodsales['kategori_UnitPrice'] = foodsales['UnitPrice'].apply(lambda x: 'mahal' if x > mean_UnitPrice else 'murah')
foodsales
```

```python
# 4
# Buatlah function untuk menghitung berapa revenue yang didapatkan di suatu kota

def revenue(nama_kota):
    city = foodsales[foodsales.City.isin(nama_kota)]
    total_revenue = city['TotalPrice'].sum()
    return total_revenue

revenue(['New York'])
```
