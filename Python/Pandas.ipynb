```python
import pandas as pd
```

```python
foodsales = pd.read_csv('FoodSales.csv')
foodsales
```

```python
# 1 
# tampilkan informasi tipe data tiap kolom

foodsales.info()
```

```python
# 2
# tampilkan semua data yang berasal dari City Los Angeles dan Quantity kurang dari 90

foodsales[(foodsales['City']=='Los Angeles') & (foodsales['Quantity'] < 90)]
```

```python
# 2
# tampilkan semua data yang berasal dari City Los Angeles dan Quantity kurang dari 90

foodsales[(foodsales.iloc[:,3]=='Los Angeles') & (foodsales.iloc[:,6] < 90)]
```

```python
# 3
# Product apakah yang paling sering dibeli di region west?

region_west = foodsales[foodsales.Region=='West']
best_seller = region_west.groupby('Product')['Quantity'].sum()
best_seller.sort_values(ascending=False).head(1)
```

```python
# 4
# Lakukan iterasi untuk membandingkan apakah nilai transaksi pada indeks tersebut lebih besar dibandingkan dengan index sebelumnya, dimulai dari index 1 (bukan 0)

foodsales['lebih_besar']

for row, col in foodsales.iterrows():
    if row > 0:
        if col['TotalPrice'] > foodsales.at[row - 1, 'TotalPrice']:
            foodsales.at[row, 'lebih_besar'] = True
print(foodsales[['TotalPrice','lebih_besar']])
```
